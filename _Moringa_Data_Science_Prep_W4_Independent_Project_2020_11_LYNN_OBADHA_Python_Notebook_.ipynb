**5. Cleaning the Dataset**

```
# Delete columns that only contain missing values: 
#
# autolib_df.dropna(axis=1, how = 'all')
autolib_df.drop('Displayed comment', axis=1, inplace=True)
```

```
autolib_df.drop('Scheduled at', axis=1, inplace=True)
autolib_df.head(3)
```

```
# Checking for and deleting duplicates to avoid redundancy.
#
autolib_df[autolib_df.duplicated()]
autolib_df.drop_duplicates(inplace = True)
```

```
# Detecting and ddeleting outliers
# Explanation: Ouliers increase variability of data, and hence decreasing the statistical power. They are removed to make results statistically significant.
#
Q1 = autolib_df.quantile(0.25)
Q3 = autolib_df.quantile(0.75)
IQR = Q3 - Q1

autolib_df_out = autolib_df[~((autolib_df < (Q1 - 1.5 * IQR)) | (autolib_df > (Q3 + 1.5 * IQR))).any(axis=1)]

print(autolib_df_out.shape)
print(autolib_df.shape)
```

```
# Combine year, month, and day columns to single datetime column.
# 
autolib_df['Date'] = pd.to_datetime(
    autolib_df[['year', 'month', 'day']].astype(str).agg('-'.join, axis=1))

autolib_df.head(7)
```

```
# Joining Hour and Minute column to time
autolib_df['time'] = (pd.to_datetime(autolib_df['hour'].astype(str) + ':' + autolib_df['minute'].astype(str), format='%H:%M')
          .dt.time)
autolib_df.head(7)
```

```
# Dropping year, month, day, hour and minute columns from the dataset
autolib_df.drop(['year', 'month', 'day', 'hour', 'minute'], axis='columns', inplace=True)
```
